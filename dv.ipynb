```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv("teen_mental_health.csv")
df
```

```python
df.head()
```

```python
df.describe()
```

```python



sns.countplot(df, x="gender", palette="Reds_r")
plt.title("Gender Distribution:")
plt.show()
```

```python

sns.histplot(df['age'],bins=5)
plt.title("Age Distribution:")
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.show()
```

```python

sns.histplot(df["daily_social_media_hours"],bins=10)
plt.title("DAILY SOCIAL MEDIA HOURS:")
```

```python

sns.boxplot(df,x="stress_level",y="daily_social_media_hours")
plt.title("SOCIAL MEDIA HOURS VS STRESS LEVEL:")
```

```python

sns.scatterplot(df,x="sleep_hours",y="mental_health_risk_score", hue="gender")
plt.title("SLEEP HOURS VS MENTAL HEALTH:")
```

```python

sns.countplot(df,x="sleep_quality")
plt.title("SLEEP QUALITY:")
```

```python

sns.countplot(df,x="depression_label")
plt.title("DEPRESSION LEVEL:")
```

```python


sns.scatterplot(df,x="academic_performance",y="mental_health_risk_score")
plt.title("ACADEMIC PERFORMANCE VS MENTAL HEALTH RISK:")
```

```python

sns.barplot(x='gender', y='sleep_hours', data=df)

plt.title("Average Sleep Hours by Gender")
```

```python
plt.figure(figsize=(10,8))

corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True,cmap='coolwarm')

plt.title("Correlation Heatmap:")
```

```python
plt.figure(figsize=(7,5))

sns.boxplot(x='physical_activity',y='mental_health_risk_score', data=df)

plt.title("Mental Health Score by Physical Activity:")
```

```python
plt.figure(figsize=(7,5))
sns.histplot(df['screen_time_before_sleep'], bins=10)

plt.title("Daily Screen Time:")
plt.xlabel("Hours")
```
